## Ответы

#### 1. В чем принципиальные отличия item-item рекомендаций от ALS?

item-item - основан на методе k ближайших соседей, поэтому у него нет процесса обучения, а только предсказание. Нет возможности регуляризации. Процесс предсказания является очень сложным.

ALS - основан на алгоритме SVD-разложения, который, в свою очередь, вычисляется методами машинного обучения (например градиентным спуском). У него есть и процесс обучения и предсказания. Есть возможность регуляризации. При этом алгоритм является довольно быстрым, особенно предсказание.

#### 3. Какие ограничения есть у ALS? (Тип информации, линейность/нелинейность факторов и т д)

- Прогнозирует не вероятность, а некие числа
- Нельзя добавлять внешние признаки
- Не учитывает сезонность

#### Сравните метрики hit_rate@k, precision@k. Какую матрику использовать предпочтительно и почему. Привидите пример 2-3 задач (опишите, что является клиентом, что товаром), в которой более уместно использовать метрику hit_rate?

hit_rate@k - определяет факт совершения покупки хотя бы одного рекомендованного товара.
precision@k - определяет долю совершенных покупок от рекомендованных.

hit_rate@k - можно использовать, если покупки очень дорогие, т.к. будет довольно странно ожидать, что пользователь, например, купит сразу несколько квартир или автомобилей.
А в остальных случаях лучше использовать precision@k.

hit_rate@k можно использовать, например, при рекомендации покупок автомобиля (клиент - покупатель, товар - автомобиль), земельный участок (клиент - покупатель, товар - земельный участок), подобрать к земельному участку дом (клиент - земельный участок, товар - дом).
