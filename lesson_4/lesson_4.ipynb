### Ответы

#### В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?

Отличие в том, что системах коллаборативной фильтрации используется только факты взаимодействия пользователя с товаром, а в гибридных системах так же учитываются дополнительные внешние признаки.

#### Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы

Примеры "холодного" старта:
- На некотором сайте интернет-продаж появился новый пользователь и нам нужно порекомендовать ему какие-то товары. Но т.к. он ещё никаких товаров в нашей системе не покупал, то для предсказания нужно использовать какие-то другие признаки, например, его пол, возраст, локацию и т.д.
- В системе видеоконтента появился новый товар (фильм) и нам нужно порекомндовать его пользователям. Но т.к. этот фильм ещё никто не смотрел, то для предсказания нужно использовать какие-то другие признаки, например, режиссёр, актёрский состав, жанр фильма и т.д.

На вебинаре рассматривалась формула: $$p = user_{emb} * item_{emb}^T + b_u + b_i$$
Судя по статье, результат этой формулы можно ещё преобразовать через некоторую функцию f, например сигмоиду. В результате в место конкретного значения мы будет получать некую вероятность интеракции пользователя с товаром.

Используются наборы данных: с плотными данными интеракций (MovieLens), и разреженными данными (CrossValidated).
- Датасет MovieLens 10M - набор данных состоит примерно из 10 миллионов оценок фильмов, присланных 71 567 пользователями 10 681 фильму. Используются теги к фильмам и показатели релевантности (от 0 до 1), обозначающим, насколько точно данный тег описывает фильм.
- Датасет CrossValidated - набор данных состоит из вопросов и ответов 5953 пользователей (44 200 вопросов и 188 865 ответов). Каждый вопрос сопровождается одним или несколькими из 1032 уникальных тегов (например, «регрессия» или «проверка гипотез»). Цель рекомендации - подобрать пользователям вопросы, на которые они могут ответить. 

LightFM сравнивается со следующими моделями:
- MF - обычная матричная модель факторизации, которую мы рассматривали на уроке 3;
- LSI-LR - модель, основанная на содержании. В отличие от LightFM, LSI-LR полностью основана на факторизации матрицы контента с использованием логистической регрессии.
- LSI-UP - гибридная модель, которая представляет профили пользователей как линейные комбинации векторов содержимого элементов, а затем применяет LSI к полученной матрице.
